```python
# Even or Odd
# Take an integer as input and print:
# "Even" if the number is even
# "Odd" otherwise
num = int(input("enter the number which you want to check: "))
if num % 2 == 0:
  print("number is even")
else:
  print("number is odd")
```

```python
# Largest of Three Numbers , Take three numbers as input and print the largest one.
num1 = int(input("enter 1st number: "))
num2 = int(input("enter 2nd number: "))
num3 = int(input("enter 3rd number: "))
print(max(num1,num2,num3))
```

```python
# Leap Year, Write a program to check whether a given year is a leap year.
year = int(input("enter the year: "))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0 ):
  print("leap year ")
else:
  print("not a leap year")
```

```python
# Sum of First N Natural Numbers
num = int(input("enter the number you have to sum: "))
sum = 0
for i in range(1,num+1):
  sum += i
print("sum = ",sum)
```
