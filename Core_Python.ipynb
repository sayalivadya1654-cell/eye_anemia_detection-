# Fstring

## Formating numbers



```python
pi = 3.14159

print(f"{pi:.2f}")   # for 2 decimal places
print(f"{pi:.4f}")   # for 4 decimal places
print(f"{pi:.5f}")   # for 5 decimal places
```

## Padding and Alignment

```python
num =  45

print(f"{num:5}")  # width →    45
print(f"{num:<5}") # left alignment → 45
print(f"{num:>5}") # right alignment →   45
print(f"{num:05}") # zero padded → 00045
print(f"{num:^5}") # centered alignment → 45
```

# Working with String

| Function       | Purpose                                       |
| -------------- | --------------------------------------------- |
| `upper()`      | Convert all letters to uppercase              |
| `lower()`      | Convert all letters to lowercase              |
| `capitalize()` | Capitalize the first letter only              |
| `title()`      | Capitalize the first letter of every word     |
| `swapcase()`   | Reverse the case of each letter               |
| `strip()`      | Remove spaces from both ends                  |
| `replace()`    | Replace part of a string                      |
| `find()`       | Find the index of a substring                 |
| `count()`      | Count occurrences of a substring              |
| `startswith()` | Check the beginning of a string               |
| `endswith()`   | Check the end of a string                     |
| `split()`      | Split a string into a list                    |
| `join()`       | Join list elements into a string              |
| `isalpha()`    | Check if all characters are letters           |
| `isdigit()`    | Check if all characters are digits            |
| `isalnum()`    | Check if all characters are letters or digits |


```python
name = "python"
print(f"{name. upper()}") # PYTHON
print(f"{name.lower()}")  # python
print(f"{name.capitalize()}") # Python
print(f"{name.title()}")  # Python
print(f"{name.swapcase()}") # PYTHON
print(f"{name.isdigit()}")  # False
print(f"{name.isalpha()}")
print(f"{name.isalnum()}")
```
