# Fstring

## Formating numbers



```python
pi = 3.14159

print(f"{pi:.2f}")   # for 2 decimal places
print(f"{pi:.4f}")   # for 4 decimal places
print(f"{pi:.5f}")   # for 5 decimal places
```

## Padding and Alignment

```python
num =  45

print(f"{num:5}")  # width →    45
print(f"{num:<5}") # left alignment → 45
print(f"{num:>5}") # right alignment →   45
print(f"{num:05}") # zero padded → 00045
print(f"{num:^5}") # centered alignment → 45
```

- extend(iterable) -> add multiple elements
- insert(i,x) -> insertat index

```python
# how to add new items in list
# append(),extend(), insert()
l = [100, 200, 300, 400, 500]
l.append(1000)  # add only one values at end of list
print(l)
l.append("string")
print(l)
l.extend([500,600,7000]) # addes multiple values in list
print(l)
l.append([1,3])
print(l)
l.extend("goa")
print(l)
l.insert(0,"sayali") # addes in desired postion in function # syntax: insert(index position, what you have to insert)
print(l)
```
