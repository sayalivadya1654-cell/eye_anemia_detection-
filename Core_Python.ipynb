## Formating numbers



```python
pi = 3.14159

print(f"{pi:.2f}")   # for 2 decimal places
print(f"{pi:.4f}")   # for 4 decimal places
print(f"{pi:.5f}")   # for 5 decimal places
```

## Padding and Alignment

```python
num =  45

print(f"{num:5}")  # width →    45
print(f"{num:<5}") # left alignment → 45
print(f"{num:>5}") # right alignment →   45
print(f"{num:05}") # zero padded → 00045
print(f"{num:^5}") # centered alignment → 45
```

# Sets {} #

- A set is a collection of unique (non-duplicate) elements.
- Sets are unordered, so items do not have a fixed position.
- Sets are mutable, meaning you can add or remove items.
- Set elements must be hashable (immutable) like numbers, - strings, and tuples.
- Sets are written using curly braces {} or the set() function

# Rules of Python Set
- Stores only unique values.
- Duplicate values are automatically removed.
- Order is not guaranteed.
- Indexing is not allowed.
- Slicing is not allowed.
- Elements must be immutable (int, float, string, tuple).
- Lists, dictionaries, and other sets cannot be stored directly.
- Can contain different data types.
- Empty set is created using set(), not {}.

# Function

- Dictionaries are mutable (can be changed).

```python
student = {
    "name": "Rahul",
    "age": 20,
    "course": "Python"
}
```

```python
# Delete
del student["course"]
```

# Functions

| **Function** | **One-Line Description**                                       |
| ------------ | -------------------------------------------------------------- |
| `len()`      | Returns the total number of key-value pairs in the dictionary. |
| `type()`     | Returns the data type of the object.                           |
| `dict()`     | Creates a new dictionary.                                      |
| `sorted()`   | Returns the dictionary keys in sorted order.                   |
| `str()`      | Converts the dictionary into a string.                         |
| `max()`      | Returns the largest key in the dictionary.                     |
| `min()`      | Returns the smallest key in the dictionary.                    |
| `all()`      | Returns `True` if all dictionary keys are true.                |
| `any()`      | Returns `True` if at least one dictionary key is true.         |


```python
# Dictionary
d = {"b": 2, "a": 1, "c": 3}

# len()
print("Length:", len(d))

# type()
print("Type:", type(d))

# dict()
new_dict = dict(name="Rahul", age=20)
print("dict():", new_dict)

# sorted()
print("Sorted Keys:", sorted(d))

# str()
print("String:", str(d))

# max()
print("Maximum Key:", max(d))

# min()
print("Minimum Key:", min(d))

# all()
print("All:", all(d))

# any()
print("Any:", any(d))
```

# Methods

| **Method**     | **One-Line Description**                                                               |
| -------------- | -------------------------------------------------------------------------------------- |
| `clear()`      | Removes all key-value pairs from the dictionary.                                       |
| `copy()`       | Returns a shallow copy of the dictionary.                                              |
| `fromkeys()`   | Creates a new dictionary with specified keys and a common value.                       |
| `get()`        | Returns the value of the specified key without causing an error if the key is missing. |
| `items()`      | Returns all key-value pairs as view objects.                                           |
| `keys()`       | Returns all the keys in the dictionary.                                                |
| `values()`     | Returns all the values in the dictionary.                                              |
| `pop()`        | Removes the specified key and returns its value.                                       |
| `popitem()`    | Removes and returns the last inserted key-value pair.                                  |
| `setdefault()` | Returns the value of a key; inserts the key with a default value if it doesn't exist.  |
| `update()`     | Updates the dictionary with key-value pairs from another dictionary or iterable.       |


```python
# Original Dictionary
d = {"name": "Rahul", "age": 20}

# copy()
copy_dict = d.copy()
print("Copy:", copy_dict)

# get()
print("Get:", d.get("name"))

# keys()
print("Keys:", d.keys())

# values()
print("Values:", d.values())

# items()
print("Items:", d.items())

# setdefault()
d.setdefault("city", "Pune")
print("After setdefault:", d)

# update()
d.update({"age": 21})
print("After update:", d)

# pop()
d.pop("city")
print("After pop:", d)

# popitem()
d.popitem()
print("After popitem:", d)

# fromkeys()
keys = ["Math", "Science", "English"]
marks = dict.fromkeys(keys, 0)
print("fromkeys:", marks)

# clear()
marks.clear()
print("After clear:", marks)
```

# control statements #

# break
- Terminates the loop immediately.
- Control jumps out of the loop as soon as break is encountered.
- Useful when a condition is met and continuing the loop is unnecessary.

# continue
- Skips the current iteration of the loop.
- Control moves directly to the next iteration without executing the remaining statements in the current cycle.
- Useful when certain values or cases should be ignored but the loop itself should keep running.

# pass
- A null operation, it does nothing.
- Acts as a placeholder where syntactically some statement is required but no action is needed.
- Often used in function definitions, class declarations, or conditional blocks during development to avoid errors while leaving the logic incomplete.

# Summary:

1. break → exits the loop entirely
2. continue → skips to the next iteration
3. pass → does nothing, placeholder

# 1. Inheritance #

- Inheritance allows one class to inherit the data members and methods of another class.
- Inheritance is the process by which one class (child class) acquires the properties (attributes) and behaviors (methods) of another class (parent class).
- Inheritance is one of the four pillars of Object-Oriented Programming (OOP).
- **super()** ➡ Calls the parent class constructor or methods.
- **Parent Class** ➡ base class/ super class
- **child class** ➡ Derived Class/ Sub Class
- Method Overriding → Child class redefines a parent method.
- **Types** → Single, Multiple, Multilevel, Hierarchical, Hybrid.
- syntax:

```python
class ParentClass:
    # Parent code
class ChildClass(ParentClass):
    # Child code
```

example:

```python
# 1. parent class
class Car: # blueprint
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model
  def info(self):
    return f"brand name : {self.brand}, model: {self.model}"
```

```python
# 2. child class
class ElectricCar(Car):
  def __init__(self, brand, model, battary_size):
    super().__init__(brand,model)
    self.battary_size = battary_size

# method
  def average(self):
    return f"about 500 as it electric car"  # Returns the common information of every car.

class FuleCar(Car):
  def __init__(self,brand, model, fule_type):
   super().__init__(brand,model)
   self.fule_type = fule_type

# Override parent method
  def info(self):
      return f"brand name : {self.brand}, model: {self.model}, fule_type: {self.fule_type}"

# method
  def average(self):
    return f"about 10 to 20 km/l, as it petrol/diseal car"
```

```python
# 3 object creation
user_ele = ElectricCar("tesla","model-s","96 kwh")
user_eng = FuleCar("toyota","corlla","petrol")
```

```python
# 4 object display
# Inheritance Output
print("----- Inheritance -----")
print(user_ele.info())
print(user_eng.info())
```

```python
# 5 display polymorphism
# Polymorphism Output
print("----- Polymorphism -----")
print(user_ele.average())
print(user_eng.average())
```

## Types of Inheritance ##
1. Single → One parent → One child.
2. Multiple → Two or more parents → One child.
3. Multilevel → Parent → Child → Grandchild.
4. Hierarchical → One parent → Multiple children.
5. Hybrid → Combination of inheritance types.
6. MRO → Defines the order Python follows to search for methods in inheritance hierarchies.


## 1. Single Inheritance ##
- Single Inheritance is a type of inheritance in which one child class inherits from one parent class.
- It is the simplest and most commonly used type of inheritance.
- Syntax :

```python
class Parent:
    # Parent code

class Child(Parent):
    # Child code
```

```python
class Animal:
   def sound(self):
      print("Animal makes sound")
class Dog(Animal):
    def bark(self):
        print("Dog barks")
obj = Dog()
obj.sound()
obj.bark()
```

## 2. Multiple Inheritance ##
- Multiple Inheritance is a type of inheritance where one child class inherits from two or more parent classes.
- The child class can access the methods and attributes of all parent classes.
- Syntax :

```python
class Parent1:
    pass
class Parent2:
    pass
class Child(Parent1, Parent2):
    pass
```

```python
class Father:
    def house(self):
        print("Father has a house")
class Mother:
    def car(self):
        print("Mother has a car")
class Child(Father, Mother):
    def bike(self):
        print("Child has a bike")
obj = Child()
obj.house()
obj.car()
obj.bike()
```

##3. Multilevel Inheritance ##
- Multilevel Inheritance is a type of inheritance where one class inherits from another child class, creating a chain of inheritance.
- syntax :

```python
class A:
    pass
class B(A):
    pass
class C(B):
    pass
```

```python
class Grandfather:
    def land(self):
        print("Grandfather owns land")
class Father(Grandfather):
    def house(self):
        print("Father owns a house")
class Son(Father):
    def bike(self):
        print("Son owns a bike")
obj = Son()
obj.land()
obj.house()
obj.bike()
```

## 4. Hierarchical Inheritance ##
- Hierarchical Inheritance is a type of inheritance where multiple child classes inherit from a single parent class.
- syntax :

```python
class Parent:
    pass
class Child1(Parent):
    pass
class Child2(Parent):
    pass
```

```python
class Animal:
    def eat(self):
        print("Animals eat food")
class Dog(Animal):
    def bark(self):
        print("Dog barks")
class Cat(Animal):
    def meow(self):
        print("Cat meows")
dog = Dog()
cat = Cat()
dog.eat()
dog.bark()
cat.eat()
cat.meow()
```

## 5. Hybrid Inheritance ##
- Hybrid Inheritance is a combination of two or more inheritance types, such as multiple inheritance and multilevel inheritance.
- Python supports hybrid inheritance, but it can become complex and often relies on MRO (Method Resolution Order) to resolve method conflicts.
- syntax:

```python
class A:
    pass
class B(A):
    pass
class C(A):
    pass
class D(B, C):
    pass
```

```python
class A:
    def displayA(self):
        print("Class A")
class B(A):
    def displayB(self):
        print("Class B")
class C(A):
    def displayC(self):
        print("Class C")
class D(B, C):
    def displayD(self):
        print("Class D")
obj = D()
obj.displayA()
obj.displayB()
obj.displayC()
obj.displayD()
```

```python
# Create a Teacher class and a MathTeacher class that inherits from it.
# Parent Class
class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject
    def info(self):
        return (
            f"Teacher Name : {self.name}\n"
            f"Subject       : {self.subject}"
        )
# Child Class
class MathTeacher(Teacher):
    def __init__(self, name, subject, experience):
        # Call Parent Constructor
        super().__init__(name, subject)
        self.experience = experience
# Create Object
teacher1 = MathTeacher("Anita Sharma", "Mathematics", "10 Years")
# Call Parent Method
print(teacher1.info())
# Access Child Attribute
print("Experience   :", teacher1.experience)
```

## Without Method Overriding ##
- The child class uses the parent class's method.
- The parent method can display only the parent's attributes.
- To display child-specific attributes, print them separately.

```python
# from above code
print(emp1.info())          # info() method used to print only Parent attributes
print(emp1.department)      # Child attribute
```

## With Method Overriding ##
- Method overriding is implemented through inheritance and is used to achieve runtime polymorphism.
- The child class creates its own method with the same name as the parent method.
- The child method can display both inherited (parent) attributes and child-specific attributes.
- No need to print child attributes separately.

```python
print(emp1.info())   # Displays emp_id, name, and department, if we add method overriding in child class
```

## 2. Polymorphism ##

- Polymorphism is one of the four pillars of Object-Oriented Programming (OOP).
- Polymorphism means **"Many Forms**."
- It allows the same method or function name to perform different tasks depending on the object.

## Types of Polymorphism ##
1. Compile-Time Polymorphism (Method Overloading)
2. Run-Time Polymorphism (Method Overriding)

###1. Compile-Time Polymorphism (Method Overloading) ###
- Method Overloading means one method can work with different numbers of arguments.
- **One method → Different arguments**
- In Python, true method overloading is not supported. We achieve a similar effect using:
1. Default arguments
2. *args

```python
# example
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

obj = Calculator()
# The same method (add) works with different numbers of arguments.
print(obj.add(10, 20))      # adds 2 numbers.
print(obj.add(10, 20, 30))  # adds 3 numbers.
```

### 2. Run-Time Polymorphism (Method Overriding) ###
- Method Overriding means a child class changes the parent's method according to its own needs.
- **One method → Different classes → Different output**
- The same method name gives different outputs depending on the object.

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")
class Dog(Animal):
    def sound(self):
        print("Dog barks")
class Cat(Animal):
    def sound(self):     # The method name is the same (sound()), but the output changes depending on the object.
        print("Cat meows")
dog = Dog()
cat = Cat()
dog.sound()
cat.sound()
```

```python
# method ovverriding
# Create a Product(name, price) class and an Electronics(name, price, warranty) class that overrides the info() method.
class Product:
  def __init__(self,name, price):
    self.name = name
    self.price = price

  def info(self):
    return f" profuct name is: {self.name}, {self.price}"
class Electronics(Product):
  def __init__(self, name, price, warranty):
    super().__init__(name,price)
    self.warranty = warranty
  def info(self):
    return (
        f"product name : {self.name} \n"
        f"product price: {self.price} \n"
        f"product warranty: {self.warranty}"
    )
customer = Electronics("washing machine","34000","5- years")
print(customer.info())
```

```python
# method overloading
# Create a Student class with a marks(math, science=0, english=0) method.
class Student:
    def marks(self, math, science= 0, english= 0):
        total = math + science + english
        return (
            f"Math Marks    : {math}\n"
            f"Science Marks : {science}\n"
            f"English Marks : {english}\n"
            f"Total Marks   : {total}"
        )
# Create Object
student1 = Student()
# Passing only Math marks
print(student1.marks(80))
# Passing Math and Science marks
print(student1.marks(80, 75))
# Passing Math, Science and English marks
print(student1.marks(80, 75, 90))
```
